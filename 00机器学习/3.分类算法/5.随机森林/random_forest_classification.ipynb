## 一、导入标准库

```python
# Importing the libraries 导入库
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# 使图像能够调整
%matplotlib notebook 
#中文字体显示  
plt.rc('font', family='SimHei', size=8)
```

# 二、导入数据

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset
```

```python
X = dataset.iloc[:, [2,3]].values # 年龄和薪水
y = dataset.iloc[:, 4].values     # 是否投放广告
X
```

```python
y
```

## 三、区分训练集和测试集

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## 四、特征缩放：决策树可以不进行特征缩放，更加直观的看到数据分类

```python
from sklearn.preprocessing import StandardScaler  # 导入标准化库
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
X_test
```

## 五、随机森林训练

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion = "entropy",n_estimators=10,random_state = 0)
classifier.fit(X_train, y_train)
```

## 六、预测测试集的结果

```python
y_pred = classifier.predict(X_test)
```

## 七、混淆矩阵

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

## 八、图形展示预测结果

### 训练集

```python
from matplotlib.colors import ListedColormap # 导入类
X_set, y_set = X_train, y_train             # 赋值变量
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green'))) # 涂色红色和绿色
plt.xlim(X1.min(), X1.max()) # 标注最大值，最小值
plt.ylim(X2.min(), X2.max()) # 标注最大值，最小值
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('orange', 'blue'))(i), label = j,s = 15)
plt.title(u'随机森林（训练集）')
plt.xlabel(u'年龄')
plt.ylabel(u'薪水')
plt.legend()
plt.show()
```

### 测试集

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('orange', 'blue'))(i), label = j,s = 15)
plt.title(u'随机森林(测试集)')
plt.xlabel(u'年龄')
plt.ylabel(u'薪水')
plt.legend()
plt.show()
```

### 1.训练集很好，测试集很烂，这个就是过拟合了。测试集的右上红色区域中没有任何黄点。。。-_-!

### 2.相比与决策树，貌似分类更精细了
